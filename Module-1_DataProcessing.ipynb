```python
all_wind = []
for i in all_json_files:
    new = pd.read_json(i)
    parts = i.split(os.path.sep)
    dummy = new["data"]["weather"]
    wind_df = pd.json_normalize(dummy, record_path = 'hourly', meta = 'date')
    wind_df['airport_code'] = parts[-2]
    all_wind.append(wind_df)
```
